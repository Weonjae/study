```python
def token(value,tokenizer=tokenizer, model=based, pool=False):
    tokens = tokenizer(value, return_tensors = "pt")
    output = model(tokens["input_ids"])
    if pool: return output[1].detach().numpy().squeeze()
    return output[0].detach().numpy()
```
